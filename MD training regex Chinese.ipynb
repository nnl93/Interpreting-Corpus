```python
import re
RE = re.compile(u'[⺀-⺙⺛-⻳⼀-⿕々〇〡-〩〸-〺〻㐀-䶵一-鿃豈-鶴侮-頻並-龎，。；：“”【】、！（）()——《》=？"^\d+$"\r\n]', re.UNICODE)
#nochinese = RE.sub('', mystring)
```

```python
import os
import nltk 
from nltk.corpus import CategorizedPlaintextCorpusReader
corpus = CategorizedPlaintextCorpusReader(
    '/users/nannanliu/Python/SCIPPC/ToRCH2014/individual/religion',
    r'(?!\.).*\.txt',
    cat_pattern=os.path.join(r'(neg|pos)', '.*',),
    encoding='utf-8')
corpus.words()
```

```python
files=corpus.fileids()
for f in files: 
    print (f)
```

```python
sub_corpora=[]
for f in files: 
    sub_corpus=corpus.raw(f)
    sub_corpora.append(sub_corpus)
```

```python
nochinese_files=[]
for sub_corpus in sub_corpora: 
    nochinese=RE.sub(' ', sub_corpus)
    nochinese_files.append(nochinese)
```

```python
for i in nochinese_files: 
    print (i)
```

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
en_words_length=[]
for file in nochinese_files: 
    en_word=tokenizer.tokenize(file)
    en_words_length.append(len(en_word))
```
