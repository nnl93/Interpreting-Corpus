```python
#count the number of words in the files, punctuations are excluded since we are dealing with interpreting oral language

def detect_language(char):
    if u'\u4e00' <= char <= u'\u9fff' or \
       u'\u3400' <= char <= u'\u4dbf':
        return 'chinese'
    elif u'\u0041' <= char <= u'\u005a' or \
         u'\u0061' <= char <= u'\u007a':
        return 'english'
    return 'other'
```

```python
import codecs
text=codecs.open('/users/nannanliu/Python/SCIPPC/2013_english_text.xml', encoding='utf-8').read()

def count_words(text):
    num_chinese_words = len([i for i in text 
                             if detect_language(i) == 'chinese'])
    english_chars = [i for i in text 
                     if detect_language(i) in ['english', 'other']]
    num_english_words = len(''.join(english_chars).split(' '))
    
    return {'Chinese': num_chinese_words, 'English': num_english_words}

words = count_words(text)

print(words)
print('Chinese words ', words['Chinese'])
print('English words ', words['English'])
print('source text word count: ', words['Chinese'] + words['English'])
```
